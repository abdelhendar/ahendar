```python
import pandas as pd
```

```python
xls = pd.ExcelFile('first.xlsx')
df1 = pd.read_excel(xls, 'Mesures')
new_header = df1.iloc[0]
df1 = df1[1:]
df1.columns = new_header
df1=df1.drop(df1.index[0], axis=0)
df1
```

```python
df1['Date']=df1['Date']+' '+df1['Heure']
df1['Date']=pd.to_datetime(df1['Date'], format='%d/%m/%Y %H:%M:%S')
```

```python
df1=df1.drop(columns=['Heure'], axis=1)
df1
```

```python
df1 = df1.groupby('Date').mean()
df1 = df1.groupby(pd.Grouper(freq="1min"))
```

```python
df1=df1.mean()
df1
```
